# disorders

```python
dis = pd.read_csv('/Users/shinsoyeon/Downloads/archive/prevalence-by-mental-and-substance-use-disorder.csv')
dis.drop(['Code'], axis=1, inplace=True)
dis.columns=['Entity', 'Year', 'Schizophrenia', 'Bipolar disorder', 'Eating disorders', 'Anxiety disorders', 'Drug use disorders', 'Depressive disorders', 'Alcohol use disorders']
dis
```

```python
dis_means=dis.iloc[:, 2:].mean(axis=1)
dis['mean']=dis_means
dis.head()
```

```python
#선진국 10개국: 미국, 중국, 러시아, 독일, 영국, 한국, 프랑스, 일본, 아랍에미리트, 이스라엘
DEV = ['China', 'Russia', 'Germany', 'United Kingdom', 'South Korea', 'France', 'Japan', 'United Arab Emirates', 'Israel']
df = dis[dis['Entity']=='United States'][['Entity', 'Year', 'mean']].reset_index()

for i in DEV:
    df_f = dis[dis['Entity']==i].reset_index()['mean']
    df = pd.concat([df, df_f], axis=1)
    
df.drop(['index', 'Entity'], axis=1, inplace=True)
df.columns=['Year', 'United States', 'China', 'Russia', 'Germany', 'United Kingdom', 'South Korea', 'France', 'Japan', 'United Arab Emirates', 'Israel']
df.head()
```

```python
#굳이 필요없는거같긴 한데 각 열을 변수로 만듦
df1=df.iloc[:,1:]

for i in range(10):
    column_name = df1.columns[i]
    column_name = column_name.replace(' ', '_')
    locals()[column_name] = df1.iloc[:, i]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=10, figsize=(20, 3))

for i, col in enumerate(df1.columns):
    df1[col].plot(ax=axes[i], title=col)
```

```python
ax = None

for col in df1.columns:
    ax = df1[col].plot(ax=ax, alpha=0.5, label=col)

plt.legend(loc='upper right')
plt.show()
```
