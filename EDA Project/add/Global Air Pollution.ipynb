# Global Air Pollution

```python
import pandas as pd
df = pd.read_csv('/Users/shinsoyeon/Downloads/global air pollution dataset.csv')
df
```

```python
df1 = df.drop(['AQI Value', 'CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value'], axis=1)
df1 = df1.replace('Good', 5).replace('Moderate', 4).replace('Unhealthy for Sensitive Groups', 3).replace('Unhealthy', 2).replace('Very Unhealthy', 1).replace('Hazardous', 0)
df1
```

```python
row = df1.iloc[:, 2:7].mean(axis=1)
df1['mean'] = row
df1
```

```python
df1.replace("United States of America", "United States", inplace=True)
df1.replace("United Kingdom of Great Britain and Northern Ireland", "United Kingdom", inplace=True)
df1.replace("Russian Federation", "Russia", inplace=True)
```

```python
import matplotlib.pyplot as plt

countries = ['United States', 'China', 'Germany', 'Japan', 'United Kingdom', 'Russia', 'France']

means = [df1[df1['Country'] == country]['mean'].mean() for country in countries]

plt.bar(countries, means, color='skyblue')
plt.xlabel('Country')
plt.ylabel('Mean Value')
plt.title('Air pollution Mean Value for Each Country')
plt.xticks(rotation=90)
plt.show()
```

# 정신건강 수치(6개 질병 평균)

```python
dis = pd.read_csv('/Users/shinsoyeon/Downloads/archive/prevalence-by-mental-and-substance-use-disorder.csv')
dis.drop(['Code'], axis=1, inplace=True)
dis.columns=['Entity', 'Year', 'Schizophrenia', 'Bipolar disorder', 'Eating disorders', 'Anxiety disorders', 'Drug use disorders', 'Depressive disorders', 'Alcohol use disorders']
dis
```

```python
dis_means=dis.iloc[:, 2:].mean(axis=1)
dis['mean']=dis_means
dis.head()
```

```python
mean = [dis[(dis['Entity']==i)&(dis['Year']==2019)]['mean'] for i in countries]

plt.plot(countries, mean, color='red')
plt.xlabel('Country')
plt.ylabel('Mean Value')
plt.title('2019 disorders by country')
plt.show()
```

```python
plt.plot(countries, mean, color='red')
plt.bar(countries, means, color='skyblue')
plt.xticks(rotation=45)
```
